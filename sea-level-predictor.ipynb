```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import linregress
%matplotlib inline
```

```python
df = pd.read_csv('sea-level.csv', na_values=['', '-', '?'])
x = df['Year']
y = df['CSIRO Adjusted Sea Level']
df
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
plt.scatter(x, y)
```

```python
result = linregress(x, y)
x1 = np.arange(df['Year'].min(),2051,1)
y1 = x1*result.slope + result.intercept
```

```python
x1
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
plt.plot(x, y, 'o', label='original data')
plt.plot(x1, y1, 'r', label='first line of best fit')
plt.legend()
plt.show()
```

```python
ndf = df[df['Year'] >= 2000]
z = ndf['Year']
v = ndf['CSIRO Adjusted Sea Level']
result2 = linregress(z, v)
x2 = np.arange(2000,2050,1)
y2 = x2*result2.slope + result2.intercept
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
plt.plot(x, y, 'o', label='original data')
plt.plot(x2, y2, 'g', label='second line of best fit')
plt.plot(x1, y1, 'r', label='first line of best fit')
plt.legend()
plt.show()
```
